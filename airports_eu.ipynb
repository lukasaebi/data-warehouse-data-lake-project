# List of the busiest airports in Europe

```python
import pandas as pd
```

```python
# URL
url = "https://en.wikipedia.org/wiki/List_of_the_busiest_airports_in_Europe"
```

```python
# Read table
tables = pd.read_html(url, header=None)

# Specifiy table
airports_table = tables[1]

print(airports_table.head(5))
```

```python
airports_table.columns = airports_table.iloc[0]  


# Reset index
airports_table.reset_index(drop=True, inplace=True)

airports_table.head()
```

```python
# Assuming airports_table is already cleaned and has correct columns
airports_table.columns = [
    "Rank", "Rank change", "Country", "Airport", "City served",
    "Passengers 2023", "Passengers 2022", "Change %", "Change People"
]

# Display the updated DataFrame
airports_table.head(5)
```

```python
# Drop the specified columns
columns_to_drop = ["Rank change", "Passengers 2022", "Change %", "Change People"]
airports_table = airports_table.drop(columns=columns_to_drop)

# Reset the index for a clean display
airports_table.reset_index(drop=True, inplace=True)
```

```python
# Remove trailing numbers in 'Country' column
airports_table['Country'] = airports_table['Country'].str.replace(r'\d+$', '', regex=True)

# Remove bracketed numbers in 'Passengers 2023' column
airports_table['Passengers 2023'] = airports_table['Passengers 2023'].str.replace(r'\[\d+\]', '', regex=True)
```
