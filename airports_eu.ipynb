# List of the busiest airports in Europe

```python
import pandas as pd
```

```python
# URL
url = "https://en.wikipedia.org/wiki/List_of_the_busiest_airports_in_Europe"
```

```python
# Read table
tables = pd.read_html(url)

# Specifiy table
airports_table = tables[1]
airports_table = airports_table.drop(index=0)
```
